```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
# plt.style.use('seaborn')
sns.set_palette("husl")

# Create data
x = np.linspace(-4, 4, 1000)
y = 1/(np.sqrt(2*np.pi)) * np.exp(-x**2/2)

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot normal distribution
ax.plot(x, y, 'b-', lw=2, label='Distribución Normal\nEstándar')
ax.fill_between(x, y, alpha=0.3)

# Add annotations for key areas
ax.annotate('68% de los datos\n(-1σ a +1σ)',
            xy=(0, 0.2),
            xytext=(0, 0.3),
            ha='center',
            arrowprops=dict(arrowstyle='->'))

# Customize plot
ax.set_title('La Distribución Normal 📊\n"La Campana que está en todas partes"', fontsize=14, pad=20)

# Add key areas annotations
areas = [
    (-1, 1, '68%', 0.25),
    (-2, 2, '95%', 0.15),
    (-3, 3, '99.7%', 0.05)
]

colors = ['#00005e', '#10f2f6', '#cfcfe1']
for (left, right, label, height), color in zip(areas, colors):
    ax.fill_between(x, y, where=(x >= left) & (x <= right), 
                    color=color, alpha=0.3)
    ax.annotate(f'{label} de los datos',
                xy=(0, height),
                ha='center',
                va='bottom')
ax.set_xlabel('Valores Estandarizados (Z-score)', fontsize=12)
ax.set_ylabel('Densidad de Probabilidad', fontsize=12)

# Add grid and legend
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()
```
